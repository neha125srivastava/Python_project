```python
import tkinter as tk
from tkinter import filedialog, messagebox

def new_file():
    text.delete(1.0, tk.END)

def open_file():
    file_path = filedialog.askopenfilename(defaultextension=".txt", filetypes=[("Text Files", "*.txt")])
    if file_path:
        with open(file_path, 'r') as file:
            text.delete(1.0, tk.END)
            text.insert(tk.END, file.read())

def save_file():
    file_path = filedialog.asksaveasfilename(defaultextension=".txt", filetypes=[("Text Files", "*.txt")])
    if file_path:
        with open(file_path, 'w') as file:
            file.write(text.get(1.0, tk.END))
            messagebox.showinfo("Info", "File saved successfully!")

# Function to apply styles
def apply_style(tag_name):
    current_tags = text.tag_names("sel.first")
    if tag_name in current_tags:
        text.tag_remove(tag_name, "sel.first", "sel.last")
    else:
        text.tag_add(tag_name, "sel.first", "sel.last")

root = tk.Tk()
root.title("Notepad")
root.geometry("900x700")

menu = tk.Menu(root)
root.config(menu=menu)
file_menu = tk.Menu(menu)
menu.add_cascade(label="File", menu=file_menu)
file_menu.add_command(label="New", command=new_file)
file_menu.add_command(label="Open", command=open_file)
file_menu.add_command(label="Save", command=save_file)
file_menu.add_separator()
file_menu.add_command(label="Exit", command=root.quit)

# Text Widget with styles
text = tk.Text(root, wrap=tk.WORD, font=("Helvetica", 12), fg="black")
text.pack(expand=tk.YES, fill=tk.BOTH)

# Adding styles
text.tag_configure("bold", font=("Helvetica", 12, "bold"))
text.tag_configure("italic", font=("Helvetica", 12, "italic"))
text.tag_configure("underline", font=("Helvetica", 12, "underline"))
text.tag_configure("red", foreground="red")
text.tag_configure("white", foreground="white")
text.tag_configure("green_bg", background="lightgreen")
text.tag_configure("purple", background="purple")

# Adding style options to the menu
style_menu = tk.Menu(menu)
menu.add_cascade(label="Style", menu=style_menu)
style_menu.add_command(label="Bold", command=lambda: apply_style("bold"))
style_menu.add_command(label="Italic", command=lambda: apply_style("italic"))
style_menu.add_command(label="Underline", command=lambda: apply_style("underline"))
style_menu.add_command(label="Red Text", command=lambda: apply_style("red"))
style_menu.add_command(label="White Text", command=lambda: apply_style("white"))
style_menu.add_command(label="Green Background", command=lambda: apply_style("green_bg"))
style_menu.add_command(label="Purple Background", command=lambda: apply_style("purple"))

root.mainloop()
```
